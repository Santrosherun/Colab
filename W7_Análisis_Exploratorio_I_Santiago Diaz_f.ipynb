```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Carga de archivo**

```python
df = pd.read_csv('https://raw.githubusercontent.com/Santrosherun/Colab/main/dataset_olympics.csv');
df
```

```python
df.head(2)
```

```python
df1 = df[['Sex', 'Age', 'Height', 'Weight', 'NOC', 'Sport', 'Medal']]
df1
```

# **Histogramas**



```python
text_dict = {'size':16,'color':'blue'};
plt.figure(figsize=(8,6));
plt.hist(df1['Height'],color='red');
plt.xlabel('Altura deportistas',fontdict=text_dict);
```

```python
text_dict = {'size':16,'color':'blue'};
plt.figure(figsize=(8,6));
plt.hist(df1['Weight']);
plt.xlabel('Peso deportistas',fontdict=text_dict);
```

```python
sns.set_style('whitegrid');

text_dict = {'size':16,'color':'blue'};
text_legend = {'size':16}
plt.figure(figsize=(8,6));
plt.hist(df1['Height'],color='red',alpha=0.4,label='Altura');
plt.hist(df1['Weight'],color='blue',alpha=0.4,label='Peso');
#plt.hist(df1['PUNT_INGLES'],color='skyblue',alpha=0.4,label='Inglés');
plt.xlabel('Relación Peso-Altura',fontdict=text_dict);
plt.legend(loc='best',prop=text_legend);
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df1,x='Height',y='Weight',hue='Sex');
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df1,x='Height',y='Age',hue='Sex');
```

```python
sns.jointplot(data=df1, x='Height',y='Age',color='b');
```

```python
sns.jointplot(data=df1, x='Height',y='Age',color='g',kind='hex');
plt.colorbar()
```

**SUBCONJUNTOS**



```python
df_basquetball = df1[df1['Sport']=='Basketball'];
df_basquetball
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df_basquetball,x='Height',y='Age',hue='Sex');
```

```python
sns.jointplot(data=df_basquetball, x='Height',y='Weight',color='g',kind='kde');
```

```python
colu = ['Height','Weight','Age'];
sns.pairplot(data=df_basquetball, x_vars=colu, y_vars=colu, hue='Sex',height=4);
```

```python
df_basquetball.columns
```

```python
sns.boxplot(data=df_basquetball,x='Sex',y='Height', showfliers=False);
```
